```python
# 디폴트 인수를 사용하는 함수 => 형식 : 변수이름 = 디폴트(기본) 값
# 함수가 호출될 때 인수가 넘어오면 넘어온 데이터로 함수를 실행하고 데이터가 넘어오지 않으면 디폴트로 지정된 데이터로 함수를
# 실행한다.
def func4(n = 100):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
```

```python
print(func4(10))
```

```python
print(func4())
```

```python
# 디폴트 인수는 부분적으로 지정할 수 없으므로 아래와 같이 함수를 선언하면 에러가 발생된다.
def func5(a = 8, b):
    print('{} + {} = {}'.format(a, b, a + b))
```

```python
def func5(a = 8, b = 6):
    print('{} + {} = {}'.format(a, b, a + b))
```

```python
func5()
```

```python
func5(9) # func5(a = 9)와 같은 표현이다.
```

```python
func5(7, 5)
```

```python
func5(b = 3) # b에만 데이터를 넘겨주고 a는 디폴트 인수를 사용한다.
```
